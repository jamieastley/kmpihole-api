```kotlin
import dev.jamieastley.kmpihole.api.KmpiholeClient
import dev.jamieastley.kmpihole.api.KmpiholeClientConfig
import dev.jamieastley.kmpihole.api.models.auth.AuthRequest
import kotlinx.coroutines.runBlocking
import kotlinx.serialization.ExperimentalSerializationApi
import kotlinx.serialization.json.Json

val client = KmpiholeClient(
    KmpiholeClientConfig(
        host = "localhost",
        port = 80,
        useTls = false,
        path = "api"
    )
)

@OptIn(ExperimentalSerializationApi::class)
val json = Json {
    prettyPrint = true
    prettyPrintIndent = "  "
}

runBlocking {
    client.auth.login(AuthRequest(password = "password"))
    client.auth.checkAuth()
}.let {
    print(json.encodeToString(it))
}

runBlocking {
    client.auth.getTotpSuggestion()
}.let {
    print(json.encodeToString(it))
}
```
