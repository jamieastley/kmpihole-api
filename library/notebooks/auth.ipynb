# Init the KmpiholeClient

Initialise the `KmpiholeClient` and `Json` instances.

**This must be executed before any other notebook blocks.**


```kotlin
import dev.jamieastley.kmpihole.api.KmpiholeClient
import dev.jamieastley.kmpihole.api.KmpiholeClientConfig
import dev.jamieastley.kmpihole.api.models.auth.AuthRequest
import kotlinx.coroutines.runBlocking
import kotlinx.serialization.ExperimentalSerializationApi
import kotlinx.serialization.json.Json
import kotlinx.serialization.json.JsonBuilder

val client = KmpiholeClient.create(
    KmpiholeClientConfig(
        host = "localhost",
        port = 80,
        useTls = false,
        path = "api"
    )
)

@OptIn(ExperimentalSerializationApi::class)
val json = Json {
    prettyPrint = true
    prettyPrintIndent = "  "
}

suspend fun login() = client.auth.login(AuthRequest(password = "password"))
```

# Log in/out


```kotlin
runBlocking {
    client.auth.checkAuth()
    client.auth.logout()
}.let {
    println(json.encodeToString(it))
}
```

# App password

```kotlin
runBlocking {
    login()
    client.auth.createAppPassword()
}.let {
    println(json.encodeToString(it))
}
```

# TOTP

```kotlin
runBlocking {
    login()
    client.auth.getTotpSuggestion()
}.let {
    println(json.encodeToString(it))
}
```
